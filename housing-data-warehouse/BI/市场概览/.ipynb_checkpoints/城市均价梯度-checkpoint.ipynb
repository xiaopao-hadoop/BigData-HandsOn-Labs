```python
import warnings
warnings.filterwarnings("ignore", message=".*pandas only supports SQLAlchemy.*")

import pymysql
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 设置绘图风格
plt.rcParams['font.sans-serif'] = ['SimHei'] 
plt.rcParams['axes.unicode_minus'] = False
sns.set_theme(style="whitegrid", font='SimHei')

def get_db_connection():
    """建立 StarRocks 数据库连接 - 严格不指定 database 参数以规避 1049 错误"""
    return pymysql.connect(
        host='192.168.144.101',
        port=9030,
        user='hadoop',
        password='hadoop',
        charset='utf8'
    )

def fetch_data():
    """
    执行 SQL 查询：获取各城市平均单价和中位数单价。
    由于建表语句未提供时间字段，默认全部分区。
    """
    conn = get_db_connection()
    # SQL 路径：catalog.db.table
    query = """
    SELECT 
        city, 
        avg_unit_price, 
        median_unit_price
    FROM iceberg_catalog.ershoufang.ads_city_overview_daily
    """
    try:
        df = pd.read_sql(query, conn)
        return df
    finally:
        conn.close()
```

```python
def plot_price_dumbbell():
    """绘制哑铃图对比均价与中位数间距"""
    # 1. 获取数据
    df = fetch_data()
    
    if df is None or df.empty:
        return

    # --- 数据准备 ---
    # 计算均价与中位数的差值，用于排序，展示分布偏斜度
    df['price_diff'] = df['avg_unit_price'] - df['median_unit_price']
    # 按差值从大到小排序，使得图表更有故事感
    df = df.sort_values(by='price_diff', ascending=True).reset_index(drop=True)

    # --- 开始绘图 ---
    fig, ax = plt.subplots(figsize=(12, 8))

    # 定义颜色和大小
    color_avg = '#D55E00'    # 朱红色 (Avg)
    color_median = '#0072B2' # 蓝色 (Median)
    
    # 1. 绘制哑铃的“杆” ( plt.hlines)
    # y轴对应排序后的城市，xmin和xmax对应两种价格
    ax.hlines(
        y=df['city'], 
        xmin=df[['avg_unit_price', 'median_unit_price']].min(axis=1), 
        xmax=df[['avg_unit_price', 'median_unit_price']].max(axis=1), 
        color='#A9A9A9',  # 杆使用灰色
        linewidth=2, 
        alpha=0.6,
        label='均价-中位数间距'
    )

    # 2. 绘制哑铃的“头” (plt.scatter - 平均值)
    ax.scatter(
        x=df['avg_unit_price'], 
        y=df['city'], 
        color=color_avg, 
        s=120,          # 点的大小
        label='平均单价', 
        zorder=3        # 确保点在杆上面
    )

    # 3. 绘制哑铃的“尾” (plt.scatter - 中位数)
    ax.scatter(
        x=df['median_unit_price'], 
        y=df['city'], 
        color=color_median, 
        s=120, 
        label='中位数单价', 
        marker='D',     # 使用菱形区分
        zorder=3
    )

    # --- 图表细节美化 ---
    plt.title('城市二手房挂牌单价：均价 vs 中位数', fontsize=16, pad=20)
    plt.xlabel('单价 (元/平米)', fontsize=12)
    plt.ylabel('城市 (按分布偏斜度排序)', fontsize=12)
    
    # 添加图例
    ax.legend(loc='lower right', frameon=True, shadow=True, title='价格类型')
    
    # 在最长的杆旁边添加解释性文字
    max_diff_idx = df['price_diff'].idxmin() # 注意排序是 ascending=True
    ax.text(
        df.loc[max_diff_idx, ['avg_unit_price', 'median_unit_price']].max() + 500, 
        df.loc[max_diff_idx, 'city'], 
        '← 间距越大，价格分布越不均匀', 
        fontsize=10, color='gray', va='center'
    )

    plt.tight_layout()
    # 首次运行 notebook 时直接展示图片
    plt.show()
```

```python
if __name__ == "__main__":
    plot_price_dumbbell()
```
